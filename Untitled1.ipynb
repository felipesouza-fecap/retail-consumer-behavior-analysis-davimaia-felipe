```python
import pandas as pd

data = pd.read_csv("/root/.cache/kagglehub/datasets/nalisha/shopping-behaviour-and-product-ranking-dateset/versions/1/shopping_behavior_updated (1).csv")

print("Primeiras linhas do dataset:")
print(data.head())

print("\nInformações gerais:")
print(data.info())

print("\nEstatísticas básicas:")
print(data.describe())

print("\nProdutos mais comprados:")
top_produtos = data["Item Purchased"].value_counts()
print(top_produtos.head(10))

print("\nCategorias mais populares:")
top_categorias = data["Category"].value_counts()
print(top_categorias)

ticket_medio = data["Purchase Amount (USD)"].mean()
print("\nTicket médio:", ticket_medio)

recomendacoes = {
    "Sneakers": "Socks",
    "T-shirt": "Jeans",
    "Dress": "Accessories"
}

def recomendar(produto):
    if produto in recomendacoes:
        return recomendacoes[produto]
    else:
        return "No recommendation available"

print("\nTestes de recomendação:")
print("Sneakers ->", recomendar("Sneakers"))
print("T-shirt ->", recomendar("T-shirt"))

data["Recommendation"] = data["Item Purchased"].apply(recomendar)

print("\nDataset com recomendações:")
print(data.head())
```
